##### Reading File

```python
#reading files 
import csv
with open('urdu-corpus.txt', 'rt', encoding="utf-8") as f:
    reader = csv.reader(f)
    passage = list(reader)
f.close()
text = passage[0][0]
print(text)
```

##### Segmentation using UrduHack
